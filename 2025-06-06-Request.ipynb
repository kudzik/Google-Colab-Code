```python
import requests

# GET - pobieranie danych
response = requests.get("https://api.github.com")
print(response.status_code, response.text)

# POST - wysy≈Çanie danych
data = {"username": "Artur", "password": "secure123"}
response = requests.post("https://example.com/login", data=data)

# PUT - aktualizacja zasob√≥w
response = requests.put("https://example.com/user/1", data={"name": "Artur"})

# DELETE - usuwanie zasob√≥w
response = requests.delete("https://example.com/user/1")
```

```python
import requests

url = "https://api.github.com"
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

response = requests.get(url, headers=headers)
# print(response)  # Wy≈õwietlenie odpowiedzi w formacie JSON

if response.status_code == 200:
    print("Sukces:", response.json())  # Pobranie danych JSON
elif response.status_code == 404:
    print("B≈ÇƒÖd: Strona nie istnieje")
elif response.status_code == 500:
    print("B≈ÇƒÖd: Problem na serwerze")
    
print("Cookies:", response.cookies)  # Wy≈õwietlenie ciasteczek zwr√≥conych przez serwer
```
